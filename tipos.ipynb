```python
# tipos
nome = 'isaac'
idade  = 20
altura = 1.80
tem_cachorro = True
nomes_familia = ['isaac', 'jose', 'maria']
perfil_isaac = {
    'nome': nome,
    'idade': idade,
    'altura': altura,
    'tem_cachorro': tem_cachorro,
    'nomes_familia': nomes_familia
}


# os tipos primitivos são:
    # string, int, float, bool, list, dict
# imprimir os tipos das variaveis
print(f"O tipo da variável 'nome' é: {type(nome)}")
print(f"O tipo da variável 'idade' é: {type(idade)}")
print(f"O tipo da variável 'altura' é: {type(altura)}")
print(f"O tipo da variável 'tem_cachorro' é: {type(tem_cachorro)}")
print(f"O tipo da variável 'nomes_familia' é: {type(nomes_familia)}")
print(f"O tipo da variável 'perfil_isaac' é: {type(perfil_isaac)}")
```
