```python
# imports
import pandas as pd
```

```python
# le um arquivo csv
df = pd.read_csv('arquivos_csv/Clientes.csv')
```

```python

# imprime os tipos das colunas
print("INFO")
df.info()
```

```python
# imprime as primeiras 5 linhas ou a quantidade que colocar no parametro
print("HEAD")
df.head(3)
```

```python
# imprime as ultimas 5 linhas ou a quantidade que colocar no parametro
print("TAIL")
df.tail(2)
```

```python
# imprime a quantidade de linhas e colunas, respectivamente
# ou seja, linhas x colunas
# linha -> df.shape[0]
# coluna -> df.shape[1]
print("SHAPE")
print(df.shape)
linhas = df.shape[0]
colunas = df.shape[1]
print(f"A tabela lida tem {linhas} linhas e {colunas} colunas.")
```

```python
# localizar apenas um item
print("LOC")
df.loc[df['state']=='Acre']
```

```python
# lcoalizar mais de um item
print("LOC por mais de um item")
states = ['Acre', 'MaranhÃ£o']
df.loc[df['state'].isin(states)]
```

```python
# localizar por condicao de numero, por exemplo, igual, maior que, menor que

# vai printar as linhas onde o id for maior ou igual a 10 e menor ou igual que 20
# usamos o & para conectar as condicoes 
df.loc[(df['id']>=10) & (df['id']<=20)]
```

```python
# selecionar colunas especificas
print("SELECIONAR COLUNAS ESPECIFICAS")
# seleciona as colunas id, first_name e number
campanha_comercial = df[['id','first_name', 'number' ,'state']]

# salva o arquivo como csv
campanha_comercial.to_csv('arquivos_csv/campanha_comercial.csv', index=False)
print("Campanha comercial salva com sucesso!")
```
