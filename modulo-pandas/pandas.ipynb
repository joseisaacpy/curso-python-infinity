```python
# filtro (dois modos de fazer)
df_estado = df.loc[df['state'] == "Acre"] # passa o estado como parametro no proprio filtro

# define uma variavel com o estado que deseja filtrar
estado_filtro = "Acre"
df_estado_outro = df.loc[df['state'] == estado_filtro] # passa a variavel como parametro e fica mais organizado
# retorna o df
df_estado_outro
```

```python
# funcao que pede um estado para filtrar
def filtro_estado(df, estado):
    # valida se existe um df
    if df is None:
        return "Dataframe nulo!"
    # valida de existe o estado no dataframe
    if not (df['state']==estado).any():
        return "Estado não encontrado!"
    else:
        df_estado_funcao = df.loc[df['state']==estado]
        return df_estado_funcao
# chama a funcao
# filtro_estado(df,"Acres")

def filtro_estado_clean(df,estado):
    filtro = df['state']==estado
    if not filtro.any():
        return "Estado não encontrado."
    return df.loc[filtro]
filtro_estado_clean(df,"Acres")
```

```python
# filtro com mais de uma opção
df_estados = df.loc[df['state'].isin(['Acre',"Paraná","São Paulo"])]
# define um array de estados
estados_filtro  = ['Acre',"Paraná","São Paulo"]
df_estados2 = df.loc[df['state'].isin(estados_filtro)]
df_estados2
```

```python
# subistituindo valores nulos
mensagem_sem_registro = 'Sem registro'
df['number'] = df['number'].fillna(mensagem_sem_registro)

df['street'] = df['street'].fillna(mensagem_sem_registro)

df['email'] = df['email'].fillna(mensagem_sem_registro)

df
```

```python
# replace - para substituir valores

# vai substituir o Brasil por Itália
df['country'] = df['country'].replace(
    to_replace="Brasil",
    value="Itália"
)
# retorna o df
df
```

```python
# replace mais avançado
mapeamento = {"Maranhão":"PI","Piauí":"PI"}
df['state'] =  df['state'].replace(mapeamento)
df
```
