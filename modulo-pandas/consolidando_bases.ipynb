```python
import pandas as pd
import os
```

```python
caminho_pasta_excel = (r'assets/arquivos_excel')

dados = []

for arquivo in os.listdir(caminho_pasta_excel):
    if arquivo.endswith('.xlsx'):
        caminho_arquivo = os.path.join(caminho_pasta_excel, arquivo)
        df = pd.read_excel(caminho_arquivo)
        dados.append(df)
consolidado = pd.concat(dados)

df = consolidado.to_csv('assets/arquivos_csv/consolidado.csv', index=False)
```
