```python
import pandas as pd
from datetime import datetime
```

```python
# pega o arquivo
caminho = 'arquivos_csv/Clientes.csv'
# faz a leitura do arquivo e retorna um dataframe
def ler_arquivo(caminho):
    df = pd.read_csv(caminho)
    return df

df = ler_arquivo(caminho)
```

```python
# faz a limpeza por colunas desnecessarias
def limpeza_colunas(df, colunas=['first_name', 'last_name', 'email','cell_phone', 'state']):
    return df[colunas]

df = limpeza_colunas(df)
```

```python
# transforma o dataframe em um csv
def salvar_csv(df, caminho):
    df.to_csv(caminho, index=False)
    print(f'Arquivo salvo em {caminho}!')

# pega a data atual
hoje = datetime.now().strftime('%d-%m-%Y')
# salva o csv
salvar_csv(df, f'arquivos_csv/base-tratada-{hoje}.csv')
```
